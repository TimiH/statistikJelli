```python
import pandas as pd
import os
#Change for future filenamechanges
filepath = '../raw/auswertung_152.' # missing extension (xslx)!
target = '../data/preprocessed/preprocessed.' # missing extension!
```

```python
#Load your dataset & save as csv(allows easier viewing and gitdiff)

df = pd.read_excel(filepath + 'xlsx')
df.to_csv(filepath + 'csv')

#Display the first few rows of the dataset
df.head()
```

```python
print(df.columns)
oldNames = df.columns
```

```python
#Rename columns for easier identification
newNames = ['time', 'check',
       'q1_1',
       'q1_2',
       'q1_3',
       'q1_4',
       'q1_5',
       'q1_6',
       'q2_1',
       'q2_2',
       'q2_3',
       'q2_4',
       'q2_5',
       'q3_1',
       'q3_2',
       'q3_3',
       'q3_4',
       'q3_5',
       'q4_1',
       'q4_2',
       'q4_3',
       'age', 'gender',
       'occupation',
       'dach']
```

```python
#Creating legend
legende = {oldNames[i]: newNames[i] for i in range(len(oldNames))}
#changing names
df = df.rename(columns = legende)
df
```

```python
# Transform ordnals scales to dummy: check,age,gender, occupation, dach

#check, 1 if 'lil', 0 otherwise
df['check'] = df['check'].replace('Lil Miquela',1)
df['check'] = df['check'].replace('Noonoouri',0)
df['check'] = df['check'].replace('Imma',0)
print(df['check'].sample(10))

#aged
#['25-34', '55-64', '64 oder älter', '18-24', '45-54', '35-44']
ageOld = df['age'].unique().tolist()
df['age'] = df['age'].astype('category').cat.codes
ageNew = df['age'].unique().tolist()
ageLegend = { ageOld[i] : ageNew[i] for i in range(len(ageOld))}
df['age'].sample(10)
# Check the preprocessing results
```
