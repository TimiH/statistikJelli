```python
import pandas as pd
import os
#Change for future filenamechanges
filepath = '../raw/auswertung_152.' # missing extension (xslx)!
target = '../data/preprocessed/preprocessed.' # missing extension!
```

```python
#Load your dataset & save as csv(allows easier viewing and gitdiff)

df = pd.read_excel(filepath + 'xlsx')
df.to_csv(filepath + 'csv')

#Display the first few rows of the dataset
df.head()
```

```python
#The column 'Ich war an dieser Werbung interessiert' needs to be removed
#It was changed mid questionaire and will be excluded

df = df.drop('Ich war an dieser Werbung interessiert',axis=1)
print(df.columns.tolist())
```

```python
#Rename columns for easier identification
oldNames = df.columns.tolist()
newNames = ['time', 'check',
       'q1_1',
       'q1_2',
       'q1_3',
       'q1_4',
       'q1_5',
       'q1_6',
       'q2_1',
       'q2_2',
       'q2_3',
       'q2_4',
       'q2_5',
       'q3_1',
       'q3_2',
       'q3_3',
       'q3_4',
       'q3_5',
       'q4_1',
       'q4_2',
       'q4_3',
       'age', 'gender',
       'occupation',
       'dach']
```

```python
#Creating legend
legende = {oldNames[i]: newNames[i] for i in range(len(oldNames))}
print(legende)
#changing names
df = df.rename(columns = legende)
```
